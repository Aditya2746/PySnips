```python
def check_no_people_know_language(x):
    return x.str.contains('Python').sum()

country_grp['LanguageWorkedWith'].apply(check_no_people_know_language).loc['India']
```

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'Number Respondents from Country', 'LanguageWorkedWith': 'Number People Know Python', 'Number People Who Know Python %' : 'Number People Who Know Python in %'}, inplace=True)
python_df
```

```python
python_df['Number People Who Don\'t Know Python'] = python_df['Number Respondents from Country'] - python_df['Number People Know Python']
python_df['Number People Who Know Python %'] = (python_df['Number People Know Python'] / python_df['Number Respondents from Country']) * 100
```

```python
python_df.sort_values(by='Number People Know Python', ascending=False, inplace=True)
```
